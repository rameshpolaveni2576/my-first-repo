**Type casting in Python**

Type casting in Python is the process of converting a variable from one data type to another, either automatically (implicit) or manually (explicit). It ensures that operations between different data types work correctly and allows programmers to control how data is interpreted.



```python
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
print(a+b)
```

```python
table_number=int(input("Enter the mul table that you want to generate: "))
for table_number in range(table_number,20,1):
  print(f"Mul table of {table_number}")
  for i in range(1,11):
    result=table_number*i
    print(f"{table_number} X {i} ={result}")
```
