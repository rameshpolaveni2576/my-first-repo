Hardware refers to the physical components of a computer system — the parts you can touch and see.
Example: CPU (Central Processing Unit) — the chip that executes instructions.

Software refers to the programs and instructions that run on the hardware — intangible code that tells the hardware what to do.
Example: Windows Operating System — it manages hardware resources and provides a user interface.
